```python
from fuzzywuzzy import process
```

```python
def correct_name(name, correct_names):
    if name in correct_names:
        return name

    new_name, score = process.extractOne(name, correct_names)
    if score < 80:
        np.append(correct_names, name)
        return name
    else:
        return new_name
```
